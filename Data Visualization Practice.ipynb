```python
#plotting using x and y coordinates and labelling axes.  If only one list then matplotlib assumes y values and generates x for U.
plt.plot([2,4,6,10], [3,5,7,9])
plt.title('Cost of Goods')
plt.ylabel('goods')
plt.xlabel('money')
plt.show()
```

```python
plt.plot([1,2,3,4])
plt.show()

'''if only one list or array specified matplotlib assumes y values and generates x for you. Since python ranges start from 0,
the default x vector has same length as y but starts with 0. Hence x data are [0,1,2,3].'''
```

```python
# Matpltlib version 2.2.2 documentation and tutorials
# for every x,y argument matplotlib accepts a third which is a format string that specifies colour and line style.
# By default the line is blue 'b-'

plt.plot([2,4,6,10], [3,5,7,9], 'ro')
plt.title('Cost of Goods')
plt.ylabel('goods')
plt.xlabel('money')
plt.axis([0,14,0,10])
plt.show()
```

```python
#More features to label and colour our plots.

x = [5,8,10]
y = [12,16,6]

x1 = [6,9,11]
y2 = [6,15,7]

plt.plot(x,y,'g', label='line 1', linewidth=5)
plt.plot(x1,y2,'c', label='line 2', linewidth=5)

plt.ylabel('Y-axis')
plt.xlabel('X-axis')
plt.title('Epic Info')

plt.legend()
plt.grid(True, color='k')
plt.show()
```
