```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline   #%matplotlib is a magic function in IPython. %matplotlib inline sets the backend of matplotlib to the 'inline' backend: With this backend, the output of plotting commands is displayed inline within frontends like the Jupyter notebook, directly below the code cell that produced it.
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/datasets/mtcars.csv')
# Using mtcars dataset from Github
```

```python
df.columns = ['cars', 'mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear', 'carb'] # Naming the columns
```

```python
# The above boxplots show stats for 3-gear, 4-gear and 5-gear cars. The 4-gear cars give more mileage
```

```python
df.boxplot(column='mpg', by='cyl')
```

```python
dd = (df.mpg - np.mean(df.mpg))/np.std(df.mpg) #This is converting to z-value to STANDARDIZE the data for mpg
```

```python
plt.plot(dd)
```

```python
dd.plot(kind='density', title='Distributon of mpg') #This yields the normal distribution for the STANDARDIZED values. Note axes!
```

```python
#What is the probability that a car produced by Mazda will provide 34 mpg?
#To compute this we need to convert to the corresponding z-score using mean of mpg.
```

```python
(34 - np.mean(df.mpg))/np.std(df.mpg)
```

```python
#Using the z score of 2.344 in a z table to find the corresponding probability of 0.4904 (49%). This is the percentage from 0.
#Half the area under the normal distribution is 50%. 
```

```python
0.50 - 0.4904 #This gives the required probability. 
```

```python
#The probability is very low because the average mpg is only around 20. 
#Thus probability of getting a car that yields 34 mpg is very low.
```
